### **1. Dataset link**

The dataset I will use for my final project is the **World Happiness Index and Inflation Dataset (2015–2023)**, available on Kaggle:
**[https://www.kaggle.com/](https://www.kaggle.com/)** (World Happiness Index and Inflation Dataset)

This dataset combines World Happiness Index scores with inflation and other macroeconomic indicators from publicly available sources, including the World Happiness Report, Gallup World Poll, and the World Bank.

---

### **2. Why is this dataset interesting? Why analyze it?**

This dataset is interesting because it links **macroeconomic conditions**, particularly inflation, with **societal well-being** measured through national happiness scores. Inflation is a central variable in business and finance, as it affects purchasing power, consumer confidence, and overall economic stability. Examining how inflation relates to happiness across countries provides insight into how economic conditions may influence quality of life.

I chose this dataset because it aligns well with my academic background in **business administration, finance, and data science**. It allows for an applied, data-driven analysis of a policy-relevant question using interpretable methods and real-world data.

---

### **3. Summary statistics (preliminary)**

After initial data cleaning, the analytical sample consists of approximately **1,200 country–year observations**.

Preliminary summary statistics show that:

* The **average happiness score** is about **5.48**, with most values falling between **4.60 and 6.33**, indicating moderate variation in reported well-being.
* **Headline consumer price inflation** averages around **7.40 percent**, with substantial dispersion across countries.
* **GDP per capita (index-scaled)** has an average value of approximately **1.03**, with less variability than inflation.

These statistics suggest that inflation varies widely across countries, making it a meaningful variable for exploring its relationship with national happiness.
