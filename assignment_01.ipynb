# Q2

```python
sales = []

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

for day in days:
    amount = float(input(f"Enter sales for {day}: "))
    sales.append(amount)

total = sum(sales)

print("Q2 Output: Total sales for the week:", total)
```

# Q3

```python
places = ["Japan", "Brazil", "Iceland", "Morocco", "Canada"]

print("Q3 Output - Original list:", places)

places.sort()
print("Q3 Output - Sorted list:", places)

places.sort(reverse=True)
print("Q3 Output - Reverse sorted list:", places)
```
